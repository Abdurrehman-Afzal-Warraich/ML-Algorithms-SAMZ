```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```
df = pd.read_csv('Salary_dataset.csv')
df
```

```
plt.xlabel('YearsExperience (years)')
plt.ylabel('Salary (US $)')
plt.scatter(df.YearsExperience,df.Salary, color='red', marker= '+')
```

```
reg = linear_model.LinearRegression()
reg.fit(df[['YearsExperience']], df.Salary)
```

```
predicted_price = reg.predict([[15]])
print("Pridicted salary is = $", predicted_price)
```

```
plt.xlabel('YearsEperience ', fontsize=20)
plt.ylabel('Salary ', fontsize=20)
plt.scatter(df.YearsExperience, df.Salary, color='red', marker='+')

# Use the correct column name 'Years Experience'
plt.plot(df.YearsExperience, reg.predict(df[['YearsExperience']]), color='blue')
```

```
d = pd.read_csv('YearsExperience.csv')
d.head(30)
```

```
input_data = np.asarray(d).reshape(-1,1)
predicted_price = reg.predict(d)

new_df = pd.DataFrame({'YearsExperience': input_data.flatten(), 'Salary': predicted_price})
new_df
```

```
# New input and its prediction
new_x = 3300
new_y = reg.predict([[new_x]])
# Plot data points
plt.scatter(df.YearsExperience, df.Salary, color='blue')

# Plot regression line
plt.plot(df.YearsExperience, reg.predict(df[['YearsExperience']]), color='blue')

# Add crossing lines for the new point
plt.axvline(new_x, color='green')
plt.axhline(new_y, color='purple')

# Plot the new point
plt.scatter(new_x, new_y, color='orange')

# Labels and legend
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression with Crossing Lines')
```
