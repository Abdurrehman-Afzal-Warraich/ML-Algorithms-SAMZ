```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('advertising.csv')
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(5, 4))
plt.xlabel('TV Advertisement')
plt.ylabel('Sales')
plt.scatter(df['TV'], df['Sales'], color='red', marker='+')
plt.show()


plt.figure(figsize=(5, 4))
plt.xlabel('Radio Advertisement')
plt.ylabel('Sales')
plt.scatter(df['Radio'], df['Sales'], color='blue', marker='o')
plt.show()

plt.figure(figsize=(5, 4))
plt.xlabel('Newspaper Advertisement')
plt.ylabel('Sales')
plt.scatter(df['Newspaper'], df['Sales'], color='green', marker='x')
plt.show()
```

```python
X=df[['TV', 'Radio', 'Newspaper']]
Y=df.Sales
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_predicted = model.predict(X)
Y_predicted
```

```python
mse = mean_squared_error(Y, Y_predicted)
mse
```

```python
r2=r2_score(Y, Y_predicted)
r2
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print("\nModel Coefficients:")
print("TV Coefficient:", model.coef_[0])
print("Radio Coefficient:", model.coef_[1])
print("Newspaper Coefficient:", model.coef_[2])
print("Intercept:", model.intercept_)
```
